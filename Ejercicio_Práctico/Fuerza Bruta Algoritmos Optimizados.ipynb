```python

PARTIDOS = [1.3, 1.3, 0.47, 1.3, 0.47, 0.75, 0.9, 0.9, 0.75, 0.9]
TV = [0.4, 0.55, 0.7, 0.8, 1, 0.45, 0.75, 0.85, 1, 0.4]


DATOS = list()
for i in range(len(PARTIDOS)):
  vista = list()
  for j in range(len(TV)):    
    vista.append(round(PARTIDOS[i]*TV[j], 3))
  DATOS.append(vista)

DATOS
```

```python
def repeticiones_partido(rep, valor):
  nuevo_valor = valor
  if rep == 2:
    nuevo_valor *= 0.75
  elif rep == 3:
    nuevo_valor *= 0.55
  elif rep == 4:
    nuevo_valor *= 0.40
  elif rep == 5:
    nuevo_valor *= 0.30
  elif rep == 6:
    nuevo_valor *= 0.25
  elif rep == 7:
    nuevo_valor *= 0.22
  elif rep == 8:
    nuevo_valor *= 0.20
  elif rep == 9:
    nuevo_valor *= 0.20
  elif rep == 10:
    nuevo_valor *= 0.20
  return nuevo_valor
```

```python
def valor(S,DATOS):
  VALOR = 0
  for i in range(len(S)):
    valor_partido  = repeticiones_partido(S.count(S[i]), DATOS[i][S[i]])
    VALOR += valor_partido
  return VALOR
```
